# **ETAPA 3: Extracción de Características y Visualización de DATOS**

## Desarrollar un análisis comparativo respecto a las 3 ciudades elegidas para el servicio Airbnb sobre las siguientes variables:

```python
#nuevas columnas llamadas "ciudad" a cada df
df_eua['ciudad'] = 'EUA'
df_mex['ciudad'] = 'MEX'
df_esp['ciudad'] = 'ESP'
```

```python
df_total = pd.concat([df_eua, df_mex, df_esp], ignore_index=True)
```

```python
df_total.info()
```

```python
from funpymodeling.exploratory import freq_tbl
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(df_total)
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(df_eua)
```

```python
freq_tbl(df_esp)
```

```python
freq_tbl(df_mex)
```

### **1) host_response_time**

```python
#Obtengo un análisis univariado de una variable categórica en específico
table_eua= freq_tbl(df_eua['host_response_time'])
table_esp= freq_tbl(df_esp['host_response_time'])
table_mex= freq_tbl(df_mex['host_response_time'])
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro1= table_eua.copy()
Filtro2= table_esp.copy()
Filtro3= table_mex.copy()
```

```python
#Ajusto el indice de mi dataframe
Filtro_index1= Filtro1.set_index('host_response_time')
Filtro_index2= Filtro2.set_index('host_response_time')
Filtro_index3= Filtro3.set_index('host_response_time')
```

```python
#Realizamos grafico de barras del dataframe filtrado
#Gráfico de barras
fig, axs = plt.subplots(1, 3, figsize=(12,4), sharey=True)
plt.subplots_adjust(wspace=0.1)

Filtro_index1["percentage"].plot(kind='bar', ax=axs[0], color= 'blue', width=0.8)
Filtro_index2["percentage"].plot(kind='bar', ax=axs[1], color= 'orange', width=0.8)
Filtro_index3["percentage"].plot(kind='bar', ax=axs[2], color= 'green', width=0.8)
axs[0].set_title('EUA')
axs[1].set_title('ESPAÑA')
axs[2].set_title('MÉXICO')

# Agregamos lineas guía en cada grafico
for ax in axs:
    ax.grid(axis='y')
    
# Agregamos títulos de los ejes
fig.text(0.08, 0.5, 'Porcentaje', va= 'center', rotation='vertical')
fig.suptitle('Tiempo que tarda un anfitrión en responder a una solicitud de reserva')
plt.show()
```

```python
# Concatenamos los DataFrames
Filtro_concat = pd.concat([Filtro_index1, Filtro_index2, Filtro_index3], axis=1, keys=['EUA', 'España', 'México'])

# Accedemos a las columnas de porcentajes para cada país
porcentajes_eua = Filtro_concat[('EUA', 'percentage')]
porcentajes_espana = Filtro_concat[('España', 'percentage')]
porcentajes_mexico = Filtro_concat[('México', 'percentage')]

# Creamos el gráfico de barras apiladas
ax = pd.concat([porcentajes_eua, porcentajes_espana, porcentajes_mexico], axis=1).plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)

# Configuramos los títulos y etiquetas
plt.title('Tiempo de respuesta por país', fontsize=14)
plt.xlabel('Tiempo de respuesta')
plt.ylabel('Porcentaje')

# Agregamos las líneas guía
ax.grid(axis='y')

# Mostramos el gráfico
plt.show()
```

### **3) number_of_reviews**

```python
df_grouped = df_total.groupby(['number_of_reviews', 'ciudad']).size().reset_index(name='conteo')
```

```python
df_grouped
```

```python
import matplotlib.pyplot as plt

# Filtramos los datos por ciudad
df_eua = df_grouped[df_grouped['ciudad'] == 'EUA']
df_mex = df_grouped[df_grouped['ciudad'] == 'MEX']
df_esp = df_grouped[df_grouped['ciudad'] == 'ESP']

# Creamos una figura con un tamaño de 10x6 pulgadas y 3 subgráficos (uno para cada ciudad)
plt.figure(figsize=(15, 5))

# Graficamos una línea para cada ciudad en su subgráfico correspondiente
plt.subplot(1, 3, 1)
plt.plot(df_eua['number_of_reviews'], df_eua['conteo'], label='EUA', color="blue")
plt.xlabel('Número de revisiones')
plt.ylabel('Conteo')
plt.title('EUA')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(df_mex['number_of_reviews'], df_mex['conteo'], label='MEX',color="green")
plt.xlabel('Número de revisiones')
plt.title('MEX')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(df_esp['number_of_reviews'], df_esp['conteo'], label='ESP',color="orange")
plt.xlabel('Número de revisiones')
plt.title('ESP')
plt.legend()

# Ajustamos el espaciado entre los subgráficos para evitar que se superpongan
plt.tight_layout()

# Mostramos el gráfico
plt.show()
```
